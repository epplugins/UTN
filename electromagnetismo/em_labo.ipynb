## Potencial de cargas puntuales

```python
np.seterr(divide='ignore')

def P_puntuales(cargas, dimension):
    # Radio mínimo:
    rm = 5
    
    # V = np.zeros(shape=(dimension,dimension), dtype=float)

    x, y = np.ogrid[0:dimension, 0:dimension]
    V = np.sqrt(np.power(x - xq, 2) + 
                    np.power(y - yq, 2))
    for carga in cargas:
        xq, yq, q = carga
        print(xq)
        D = np.sqrt(np.power(x - xq, 2) + 
                    np.power(y - yq, 2))
        V = V + 9E9 * 1000 * q / D

    return V

def graf_Vpuntuales(V, x1, y1, h, figsize=(6,6), titulo='Título'):
    # dx, dy = V.shape
    # xaxis, yaxis = np.ogrid[0:dx, 0:dy]
    x2 = x1 + h
    y2 = y1 + h

    x = np.arange(x1, x2)
    y = np.arange(y1, y2)
    X, Y = np.meshgrid(x, y)
    fig, ax = plt.subplots(1, 1, figsize=figsize,facecolor=(1, 1, 1) )
    # CS2 = ax.contour(X, Y, np.transpose(D)[x1:x2,y1:y2])
    CS2 = ax.contour(X, Y, np.transpose(V)[x1:x2,y1:y2])
    ax.clabel(CS2, inline=True, fontsize=10)
    ax.set_title(titulo)
    plt.show()
    return True
```

```python
cargas = [
    [300, 200, 2.67E-11],
    [400, 300, -2.67E-11],
]
P = P_puntuales(cargas, 1000)
xr = 300
yr = 200
h = 100
out = graf_Vpuntuales(P, xr, yr, h)
```
